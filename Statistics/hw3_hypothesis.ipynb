Да, проблемы есть. Это выбросы. Хвосты на гистограммах оказываются очень длинными. Найдите медианы и средние для числа лайков. Сравните их между собой. 

```python
### ╰( ͡° ͜ʖ ͡° )つ▬▬ι═══════  bzzzzzzzzzz
# your code below
df_likes_true.describe()[['mean', '50%']]
```

Проверьте гипотезу о равенстве среднего числа лайков. Используйте для этого $z$-тест, основанный на ЦПТ. Дисперсию расчитывайте по выборке, используя несмещённую оценку `(ddof=1)`. Запишите наблюдаемое значение статистики в переменную `z_obs`.  
$H_0:$ `likes true mean` = `likes false mean`  
$H_1:$ `likes true mean` != `likes false mean`

$$
z=\dfrac{\bar{x}-\bar{y}}{\sqrt{\dfrac{\hat{\sigma}_{x}^{2}}{n_x}+\dfrac{\hat{\sigma}_{y}^{2}}{n_y}}} \sim N(0,1)
$$

```python
### ╰( ͡° ͜ʖ ͡° )つ▬▬ι═══════  bzzzzzzzzzz
# your code below

from scipy import stats

alpha = 0.05
n_true, n_false = df_likes_true.size, df_likes_false.size             # размер выборок
likes_true = df_likes_true.mean()
likes_false = df_likes_false.mean()
diff_likes = likes_true - likes_false             # разность средних

diff_likes_std = np.sqrt(df_likes_true.var(ddof=1)/n_true + df_likes_false.var(ddof=1)/n_false)   # стандартное отклонение

z_obs = diff_likes/diff_likes_std
z_obs
```

Найдите критическое значение статистики. Сравните его с наблюдаемым. Рассчитайте p-value для ситуации с двусторонней альтернативой. Сделайте вывод о выдвинутой гипотезе. Критическое значение статистики запишите в переменную `z_crit`, а p-значение в переменную `pval`. 

```python
### ╰( ͡° ͜ʖ ͡° )つ▬▬ι═══════  bzzzzzzzzzz
# your code below
z_crit = sts.norm.ppf(1-alpha/2)
pval = sts.norm.sf(abs(z_obs))*2
if pval <= alpha:
  print(round(pval, 3), 'Reject the null hypothesis: Mean_I_True = Mean_I_False')
else:
  print(round(pval, 3), 'Do not reject the null hypothesis: Mean_I_True = Mean_I_False')
```

```python
# проверка того, что задание решено верно
assert np.abs(z_obs - 2.6644610922505323) < 1e-5
```

Можно заметить, что у распределения числа репостов длинные хвосты. В данных есть выбросы, которые завышают среднее значение. Если выбросы сильные, то ЦПТ может ломаться. Из-за этого имеет смысл в данном случае проверять гипотезу о равенстве медиан, которые устойчивы к выбросам.

## Тайлер

__Люди всё время меня спрашивают, знаю ли я Тайлера Дёрдена?__ Да, знаю. Он делает лучшее мыло в стране. 

__а)__ Тайлер считает партию изделий хорошей, если дисперсия размера не превышает $25$ грамм$^2$. Пока Тайлер летел в самолёте, он измерил $10$ кусков мыла. Выборочная дисперсия оказалась равна $27$ грамм$^2$. Можно ли продавать эту партию или её надо списать в брак? 

При проверке этой гипотезы найдите $p$-значение и запишите его в переменную `pval1`. Наблюдаемое значение статистики запишите в переменную `chi_obs`. Используйте уровень значимости $1\%$. [Chi-squared test](https://www.itl.nist.gov/div898/handbook/eda/section3/eda358.htm)  
$H_0:$ `sample var` = `population var`  
$H_1:$ `sample var` > `population var` 

```python
### ╰( ͡° ͜ʖ ͡° )つ▬▬ι═══════  bzzzzzzzzzz
# your code below
alpha = 0.01
N = 10
s = 27**0.5
sigma = 25**0.5
chi_obs = (N-1)*(s/sigma)**2
pval1 = sts.chi2.ppf(alpha, N-1)
if pval1 <= alpha:
  print(round(pval1, 3), 'Reject the null hypothesis: sample_var = pop_var')
else:
  print(round(pval1, 3), 'Do not reject the null hypothesis: sample_var = pop_var')
```

__б)__ Для того, чтобы уменьшить количество бракованного мыла, Тайлер собирается внедрить новую технологию производства. Тайлер произвёл по старой технологии $100$ кусков мыла. Стандартное отклонение этой партии составило $6$ грамм. Для  новой технологии стандартное отклонение, посчитанное по $100$ кускам мыла, составило $5$ грамм. Правда ли, что новая технология обладает меньшей дисперсией и её нужно внедрить на производстве?

При проверке этой гипотезы найдите $p$-значение и запишите его в переменную `pval2`. Используйте уровень значимости $1\%$. [F-Test](https://www.itl.nist.gov/div898/handbook/eda/section3/eda359.htm)  
$H_0:$ `new var` = `old var`  
$H_1:$ `new var` < `old var`

```python
### ╰( ͡° ͜ʖ ͡° )つ▬▬ι═══════  bzzzzzzzzzz
# your code below
alpha = 0.01
N = 100
s = 6
sigma = 5
F_obs = s**2/sigma**2 
pval2 = sts.f.ppf(1-alpha, N-1, N-1)
if pval2 <= alpha:
  print(round(pval2, 3), 'Reject the null hypothesis: new_var = old_var')
else:
  print(round(pval2, 3), 'Do not reject the null hypothesis: Mean_I_True = Mean_I_False')
```

```python
# проверка того, что задание решено верно
assert np.abs(chi_obs - 9.72) < 1e-1
assert np.abs(F_obs - 1.44) < 1e-1
```

## Рейтинги президентов

В таблице `presidents.csv` представлены рейтинги различных президентов США. Социальные опросы проводились раз в квартал. Опрашивалась случайная выборка граждан. У каждого спрашивалось одобряет ли он деятельность президента.

```python
df = pd.read_csv('/content/drive/MyDrive/Math & Natural & Finance/Math/statistics/data/presidents.csv')
df.head()
```

Чтобы привести таблицу к удобному виду используйте функцию `pd.melt`. Вытащите из таблички рейтинг Никсона с 1 квартала 1969 года по 2 квартал 1972 года (включая его). Запишите эти данные в переменную `x`. Рейтинг с 3 квартала 1972 года по 2 квартал 1974 года поместите в переменную `y`. Пропуски в данных удалите из выборки.

```python
### ╰( ͡° ͜ʖ ͡° )つ▬▬ι═══════  bzzzzzzzzzz
# your code below
df_1 = pd.melt(df, id_vars=['year'], value_vars=['Qtr1', 'Qtr2', 'Qtr3', 'Qtr4'])
df_1['time'] = df_1['year'] + df_1['variable'].str[-1].astype(int)/10

x = df_1.query('1969<=time<=1972.2')['value'].dropna().values #df_1[df_1['time'].apply(lambda x: 1969 <= x <= 1972.2)]['value'].dropna().values
y =  df_1.query('1972.3<=time<=1974.2')['value'].dropna().values #df_1[df_1['time'].apply(lambda x: 1972.3 <= x <= 1974.2)]['value'].dropna().values 
```

Правда ли, что после скандала средний рейтинг Никсона рухнул? Предполагайте, что рейтинг распределён нормально, а также, что дисперсия рейтинга не изменилась из-за скандала. Наблюдаемое значение статистики запишите в переменную `obs`, в переменную `pval` запишите $p$-значение.  
$H_0:$ `before mean` = `after mean`  
$H_1:$ `before mean` != `after mean`

Ниже отдельно посчитана [вариация](https://tsput.ru/res/informat/mop/lections/lection_6.htm) для двух разных выборок, имеющих одинаковую дисперсию:

```python
x_2 = (x - x.mean())**2
y_2 = (y - y.mean())**2
xy_2 = (np.sum(x_2)+np.sum(y_2))/(x.size + y.size - 2)
```

```python
### ╰( ͡° ͜ʖ ͡° )つ▬▬ι═══════  bzzzzzzzzzz
# your code below

alpha = 0.05
n_1, n_2 = x.size, y.size             # размер выборок
mean_1 = x.mean()
mean_2 = y.mean()
diff_mean = mean_1 - mean_2             # разность средних
diff_std = np.sqrt(xy_2/n_1 + xy_2/n_2)   # стандартное отклонение

obs = diff_mean/diff_std
print(obs)

crit = sts.norm.ppf(1-alpha) # допущение о нормальном распределении
pval = sts.norm.sf(abs(obs))*2
if pval <= alpha:
  print(round(pval, 5), 'Reject the null hypothesis: before_mean = after_mean')
else:
  print(round(pval, 5), 'Do not reject the null hypothesis: before_mean = after_mean')
```

```python
# проверка того, что задание решено верно
assert np.abs(obs - 3.5640741313229793) < 1e-5
```

Проводится эксперимент о пользе нового снотворного. Для его проведения используется группа из $10$ студентов. В первый день каждый из них ложится спать. На утро измеряется, сколько он проспал. Из результата вычитают $8$ (нормальное число часов сна). На второй день каждому дают таблетку. Утром снова измеряют сколько человек проспал, и вычитают $8$. Результаты эксперимента записаны в табличку `sleep.csv`.

```python
df = pd.read_csv('/content/drive/MyDrive/Math & Natural & Finance/Math/statistics/data/sleep.csv')
df.head()
```

Правда ли, что сон студентов до приёма снотворного не отличается от нормы (отклонение от 8 часов равно нулю). Проверьте эту гипотезу на $5\%$ уровне значимости, в переменную `pval1` запишите p-значение.   
$H_0:$ `group 1 mean` = 0  
$H_1:$ `group 1 mean` != 0 

```python
### ╰( ͡° ͜ʖ ͡° )つ▬▬ι═══════  bzzzzzzzzzz
# your code below

#samples
gr_1 = np.array(df[df['group'] == 1]['extra'])
#sample means
mean_1 = gr_1.mean()
#sample vars (ddof = 1)
var_1 = gr_1.var(ddof=1)

tobs1 = (mean_1-0)/(var_1/gr_1.size)**0.5
pval1 = sts.t(df=gr_1.size).sf(abs(tobs1))*2
crit1 = sts.t(df=gr_1.size).ppf(1-alpha/2)

if tobs1 >= crit1:
  print(round(pval1, 5), 'Reject the null hypothesis: group 1 mean = 0')
else:
  print(round(pval1, 5), 'Do not reject the null hypothesis: group 1 mean = 0')
```

На уровне значимости $5\%$ гипотеза о неотличимости продолжительности сна **отвергается**

Правда ли, что благодаря таблеткам студенты стали спать больше (односторонняя альтернатива)? Запишите в переменную `pval2` соотвествующее p-значение. [t-test for 2 dependent samples](http://srjcstaff.santarosa.edu/~ylin/Math15/notes/Compare_Two_Dependent_Samples.htm)  
$H_0:$ `gr_2` = `gr_1`  
$H_1:$ `gr_2` > `gr_1`

```python
### ╰( ͡° ͜ʖ ͡° )つ▬▬ι═══════  bzzzzzzzzzz
# your code below
d = np.array(df[df['group'] == 2]['extra']) - np.array(df[df['group'] == 1]['extra'])
d.std(ddof=1)
tobs2 = (d.mean() - 0)/(d.std(ddof=1)/d.size**0.5)
pval2 = sts.t(df=d.size-1).sf(abs(tobs2))*2
crit2 = sts.t(df=d.size-1).ppf(1-alpha/2)

if pval2 <= alpha:
  print(round(pval2, 5), 'Reject the null hypothesis: Mean difference = 0')
else:
  print(round(pval2, 5), 'Do not reject the null hypothesis: Mean difference = 0')
```

- Вытащим по аналогии информацию о том, откликнулась ли кошка на своё имя (пометка `Test Cat's Name`) в колонке `stimuli`. Запишем её в вектор `name`.
- Используем тест о равенстве пропорций для зависимых выборок, чтобы ответить на вопрос "А правда ли кошка откликается на своё имя?" В переменную `pval` запишем $p$-значение, полученное при проверке этой гипотезы. 

$H_0:$ `name` = `noun_4`  
$H_1:$ `name` > `noun_4`

```python
#noun4
df_1 = df[df['stimuli'] == 'Noun 4'].reset_index(drop=True)
df_1['bin_answer'] = np.where(df_1['Total Score'] > 0, 1, 0)
df_1['bin_answer'].head() 
```

```python
#name
df_2 = df[df['stimuli'] == 'Test Cat\'s Name'].reset_index(drop=True)
df_2['bin_answer'] = np.where(df_2['Total Score'] > 0, 1, 0)
df_2['bin_answer'].head()
```

```python
### ╰( ͡° ͜ʖ ͡° )つ▬▬ι═══════  bzzzzzzzzzz
# your code below
alpha = 0.05

noun4 = df_1['bin_answer'] 
name = df_2['bin_answer']

#number of matches 
to_name = ((np.subtract(name, noun4)>0)*np.subtract(name, noun4)).sum()
to_noun4 = abs(((np.subtract(name, noun4)<0)*np.subtract(name, noun4)).sum())

z_obs = (to_name-to_noun4)/(to_name+to_noun4-(to_name-to_noun4)**2/noun4.size)**0.5
pval = sts.norm.sf(abs(z_obs)) #one-tailed test: H_1: p_name > p_noun4
z_crit = sts.norm.ppf(1-alpha)

if pval <= alpha:
  print(round(pval, 5), 'Reject the null hypothesis: p_name = p_noun4')
else:
  print(round(pval, 5), 'Do not reject the null hypothesis: p_name = p_noun4')
```

Лекции по матану на экономическом факультете читают два преподавателя: Злой и Добрый. Каждый год они чередуются. Среди студентов ходит слух, что в год, когда преподаёт Злой, отчисляют куда больше народу. В таблице `matan.csv` лежит информация по $418$ студентам экономического факультета. 

В столбце `is_angry` стоит единица, если этому студенту преподавал Злой. В столбце `target_1` стоит единица, если человек закончил первый курс. Проверьте гипотезу о том, что Злой отчисляет больше народу. В переменную `pval` запишите $p$-значение, получившееся при проверке гипотезы. 

$H_0:$ `p_angry` = `p_good`  
$H_1:$ `p_angry` > `p_good`

```python
df = pd.read_csv('/content/drive/MyDrive/Math & Natural & Finance/Math/statistics/data/matan.csv', sep='\t')
df = df.dropna(axis=0)
df.shape #.head()
```

```python
### ╰( ͡° ͜ʖ ͡° )つ▬▬ι═══════  bzzzzzzzzzz
# your code below
alpha = 0.05

angry = df[df['is_angry'] == 1]['target_1'].reset_index(drop=True) 
good = df[df['is_angry'] == 0]['target_1'].reset_index(drop=True) 

#number of matches 
sum_angry = sum(angry)
sum_good = sum(good)

#estimated frequences 
p_angry_hat = sum_angry/angry.count()
p_good_hat = sum_good/good.count()
assert (p_angry_hat <= 1)&(p_good_hat <= 1)

#proportion of "successes" in the two samples combined 
p_combined = (sum_angry+sum_good)/(angry.count()+good.count())

z_obs = (p_angry_hat-p_good_hat-0) /  (p_combined*(1-p_combined)*(1/angry.count()+1/good.count()))**0.5 #(p_angry_hat*(1-p_angry_hat)/angry.count() + p_good_hat*(1-p_good_hat)/good.count())**0.5
pval = sts.norm.sf(abs(z_obs))
z_crit = sts.norm.ppf(1-alpha)

if pval <= alpha:
  print(round(pval, 5), 'Reject the null hypothesis: p_angry = p_good')
else:
  print(round(pval, 5), 'Do not reject the null hypothesis: p_angry = p_good')
```

## Энергетик

Исследовательница Глафира считает, что любовь к энергетическим напиткам и успешность сдачи экзамена по математической статистике должны быть как-то связаны. Опросив $200$ своих однокурсников, она получила следующие результаты:

- Сдал, пьёт энергетик: 20
- Сдал, не пьёт энергетик: 120
- Завалил, пьёт энергетик: 40
- Завалил, не пьёт энергетик: 20

Правда ли, что энергетик оказывает на успешность сдачи экзамена какое-то влияние? Проверьте гипотезу с помощью критерия Пирсона. В качестве ответа запишите в переменную `pvalue` получившееся $p$-значение.

$
\begin{aligned}
H_0 : X \sim F_0 \\
H_1 : X \nsim F_0
\end{aligned}
$

```python
### ╰( ͡° ͜ʖ ͡° )つ▬▬ι═══════  bzzzzzzzzzz
# your code below
alpha = 0.05
n = 200
observed_cnts = np.array([20, 120, 40, 20])
p = observed_cnts/n # observed frequences
#print(p)

expected_prob = np.array([1/4]*4) #assuming equal probability of the each scenario
#print(expected_prob)

# Pearson's statistics
T = np.sum((observed_cnts - n*expected_prob)**2 / (n*expected_prob))
#print(T)

chi2_crit = stats.chi2.ppf(1-alpha, df=4-1)
#print(chi2_crit)

pvalue = 1 - stats.chi2.cdf(T, df=4-1) 

if pvalue <= alpha:
  print(round(pvalue, 5), 'Reject the null hypothesis: x = F_0')
else:
  print(round(pvalue, 5), 'Do not reject the null hypothesis: x = F_0')
```

```python
# проверка того, что задание решено верно
assert pvalue < 0.05
```

После терапии тоже.

Поэтому ассмотрим, для начала, тест на нормальность специфический к тяжёлым хвостам [(Андерсона-Дарлинга)](https://www.statsmodels.org/stable/generated/statsmodels.stats.diagnostic.normal_ad.html#statsmodels.stats.diagnostic.normal_ad):

```python
from statsmodels.stats.diagnostic import normal_ad
alpha = 0.05
res = normal_ad(df[['Before', 'After']])

if res[0][1] <= alpha:
  print(round(res[0][1], 5), 'Reject the null hypothesis: F_Before = Norm')
else:
  print(round(res[0][1], 5), 'Do not reject the null hypothesis: F_Before = Norm')

if res[1][1] <= alpha:
  print(round(res[1][1], 5), 'Reject the null hypothesis: F_After = Norm')
else:
  print(round(res[1][1], 5), 'Do not reject the null hypothesis: F_After = Norm')
```

На уровне значимости 5% выборки из **нормального распределения**.

Проверим с помощью теста Колмогорова-Смирнова, что выборки пришли из одинакового распределения на уровне $\alpha = 5 \%$:

```python
alpha = 0.05

pvalue1 = sts.ks_2samp(df['Before'], df['After'])[1]

if pvalue1 <= alpha:
  print(round(pvalue1, 5), 'Reject the null hypothesis: F_0 = F_1')
else:
  print(round(pvalue1, 5), 'Do not reject the null hypothesis: F_0 = F_1')
```

**Параметрический тест:** Так как мы сравниваем эффект от терапии на одних и тех людях (зависимые выборки), эффективность имеет смысл проверить *как разность средних для зависимых выборок*: [Paired t-test](http://srjcstaff.santarosa.edu/~ylin/Math15/notes/Compare_Two_Dependent_Samples.htm).

1) Найдём изменение веса для каждого пациента:

```python
df['d'] = df['After'] - df['Before']
```

2) Посчитаем дисперсию и среднее приращения:

```python
mu_d = df['d'].mean()
var_d = df['d'].var(ddof=1)
```

3) Сформулируем гипотезу:  
$H_0\colon \mu(A) = \mu(B)$

$H_1\colon \mu(A) > \mu(B)$ 

```python
alpha = 0.05

t_obs = (mu_d-0)/(var_d/df.shape[0])**0.5
pval = sts.t(df=df.shape[0]-1).sf(abs(t_obs))
t_crit = sts.t(df=df.shape[0]-1).ppf(1-alpha)

if pval <= alpha:
  print(round(pval, 5), 'Reject the null hypothesis: mu_B = mu_A')
else:
  print(round(pval, 5), 'Do not reject the null hypothesis: mu_B = mu_A')
```

**Непараметрический тест:** Критерии знаков (двухвыборочный) и Критерии Уилкоксона (двухвыборочный). Тесты можно использовать, т.к выборки из одного распределени (на 5%) и связанные.

Критерий знаков  
$H_0\colon P(After > Before) = 0.5$

$H_1\colon P(After > Before) \ne 0.5$

```python
# Критерий знаков
from statsmodels.stats.descriptivestats import sign_test

x = df.After
y = df.Before
_, pvalue = sign_test(x, y)

if pvalue <= alpha:
  print(round(pvalue, 5), 'Reject the null hypothesis: P(After > Before)')
else:
  print(round(pvalue, 5), 'Do not reject the null hypothesis: P(After > Before)')
```

Критерий Уилкоксона  
$H_0\colon Med(After > Before) = 0$

$H_1\colon Med(After > Before) \ne 0$

```python
_, pvalue = sts.wilcoxon(x, y)

if pvalue <= alpha:
  print(round(pvalue, 5), 'Reject the null hypothesis: Med(After > Before)')
else:
  print(round(pvalue, 5), 'Do not reject the null hypothesis: Med(After > Before)')
```

**Итог:** На уровне значимости 5% терапия даёт эффект.

Параметрический тест и Критерий знаков подтвердили эффективность, а критерий Уилкоксона нет.

**Параметрический тест:** Проверим гипотезу об изменении цены:  
$H_0\colon \mu(2001) = \mu(2002)$

$H_1\colon \mu(2001) < \mu(2002)$ 

* Объекты хоть и из одного района, но не одни и те же, поэтому выборки не рассматриваю их как зависимые выборки.  
* Дисперсии неизвестны - воспользуемся распределением Уэлча:

```python
alpha = 0.05                                                    # уровень значимости = 5%
n_2002, n_2001 = df_0102['2002'].size, df_0102['2001'].size                                   # размер выборок

diff_price = df_0102['2002'].mean() - df_0102['2001'].mean()                                 # разность средних
diff_price_std = np.sqrt(df_0102['2002'].var(ddof=1)/n_2002 + df_0102['2001'].var(ddof=1)/n_2001)   # стандартное отклонение

# число степеней свободы
num = (df_0102['2002'].var(ddof=1) / n_2002 + df_0102['2001'].var(ddof=1) / n_2001) ** 2
den = (df_0102['2002'].var(ddof=1)**2) / (n_2002**2 * (n_2002 - 1)) + (df_0102['2001'].var(ddof=1)**2)/ (n_2001**2* (n_2001 - 1))

t_obs = diff_price/diff_price_std
pval1 = sts.t(df=num/den).sf(abs(t_obs))
crit1 = sts.t(df=num/den).ppf(1-alpha)
```

```python
if t_obs >= crit1:
  print('delta price =', int(diff_price), ', p-value =', round(pval1, 5), 'Reject the null hypothesis: mu_2001 = mu_2002')
else:
  print('delta price =', int(diff_price), ', p-value =', round(pval1, 5), 'Do not reject the null hypothesis: mu_2001 = mu_2002')
```

**Непараметрический тест:** из-за несвязности выборок (вообще говоря), можно воспользоваться Критерием Манна-Уитни (двухвыборочный). Таким образом покажем (или нет), что выборки имеют одно распределение, а сдвиг между ними - это и есть изменение цены.  

$H_0\colon f_x(x) = f_y(y)$

$H_1\colon f_x(x) = f_y(y+\Delta), \space \Delta \neq 0$

```python
from scipy.stats import mannwhitneyu
alpha = 0.05
U1, p = mannwhitneyu(df_0102['2001'], df_0102['2002'], alternative = 'less')
if p <= alpha:
  print('p-value =', round(p, 5), 'Reject the null hypothesis: f_2001 = f_2002')
else:
  print('p-value =', round(p, 5), 'Do not reject the null hypothesis: f_2001 = f_2002')
```

```python
# по неизвестному распределению выделим много-много выборок, которые в среднем будут дадут истинное распределение
def get_bootstrap_sample(x, B_sample=1):
    N = x.size
    sample = np.random.choice(x, size=(N, B_sample), replace=True)
    
    if B_sample == 1:
        sample = sample.T[0]
    return sample

def stat_intervals(boot, alpha=0.05):
    lower = np.quantile(boot, alpha/2)
    upper = np.quantile(boot, 1-alpha/2)
    return lower, upper
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df[['vs', 'am']], 
             x='vs', 
             hue='am', 
             bins=10,
             common_norm=False, 
             stat='density',
            );
```

По бинарной диаграмме сложнее делать выводы, но тенденция есть: V-образному двигателю чаще соответствует автомат; рядному - механика.

Для категорийных переменных применяется [Тест Фишера](https://stats.stackexchange.com/questions/408564/relationship-independence-between-2-binary-varibales), [scipy](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.fisher_exact.html) (или [$\chi^2$](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.chi2_contingency.html#scipy.stats.chi2_contingency) \):

```python
a = df[(df['vs'] == 0)&(df['am'] == 0)].shape[0]
b = df[(df['vs'] == 0)&(df['am'] == 1)].shape[0]
c = df[(df['vs'] == 1)&(df['am'] == 0)].shape[0]
d = df[(df['vs'] == 1)&(df['am'] == 1)].shape[0]
res = [[a, b], [c, d]]
res
```

Из полученной таблицы вытекает, что V-образных моторов на автомате больше (12), чем на механике (6).

```python
from scipy.stats import fisher_exact
oddsr, p = fisher_exact(res, alternative='two-sided')
p
```

`p-value` >> 5%, т.е наша смещение нашей выборки статистически значимо (гипотеза о влиянии не отвергается) и с вероятностью 47% и более мы можем увидеть ещё более смещённую выборку: таким образом, АКПП чаще ставится на V-моторы, а механика на рядные.
