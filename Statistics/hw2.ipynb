Неравенство Чёбышева для вероятности отклонений: $P(|X-M(X)| \geq k \sigma) \leq \frac{1}{k^2} ⇔  
P(|X-M(X)| < k \sigma) > 1 - \frac{1}{k^2}$.  
* При k = 2 имеем: $P(|X-a| < 2 \sigma) > \frac{3}{4} $ (~0.75);
* При k = 3 имеем: $P(|X-a| < 3 \sigma) > \frac{8}{9}$ (~0.889).

```python
k_2 = round(2*stats.norm(0, 1).cdf(2)-1, 4)
k_3 = round(2*stats.norm(0, 1).cdf(3)-1, 4)
print('При k = 2: P = {0}. При k = 3: P = {1}.'.format(k_2, k_3))
```

n1) Неравенство Чебышева даёт более грубую оценку нахождения с.в внутри определённого интервала. поэтому можно заметить, что для произвольного распределения вероятность для $2\sigma$ и $3\sigma$ сильно меньше, чем таковая для нормального распределения: 0.750 vs. 0.954 и 0.889 vs. 0.997.

# Равномерная случайная величина

Рассмотрим случайную величину из равномерного распределения на отрезке от $0$ до $θ$ с функцией распределения:

$$
F(x) = \begin{cases} 0, & x < 0 \\ x/\theta, & 0 \le x \le \theta \\ 1, & x > \theta. \end{cases}
$$

и с плотностью распределения:

$$
f_X(x) = \begin{cases} 1/\theta, & 0 \le x \le \theta \\ 0, & \text{otherwise.} \end{cases}
$$

или через индикаторную функцию:

$$
f_X(x) = \frac{1}{\theta} \cdot \mathbb 1\{0 \le x \le \theta\}.
$$

- состоятельными $ \hat{\theta} \overset{\mathbb{P}}{\to} \theta , n\to ∞$

Используйте тот факт, что распределение максимума из $n$ независимых равномерно распределённых на отрезке $[0, θ]$ случайных величин имеет функцию распределения:

$$
\mathbb{P}\left(\max \left(X_{1}, \ldots, X_{n}\right)<x\right)=[\mathbb{P}(X_i<x)]^{n}=\left(\frac{x}{\theta}\right)^{n}=F(x)=\text{cdf}(\hat{\theta}_1)
$$

**n1) Покажите, что оценка параметра $θ$ метом максимального правдоподобия есть оценка $\hat{\theta}_1 = \max(X_1, ..., X_n)$**  

Функция правдободобия будет: $L = f(\theta) = \prod_{j=1}^n \frac{1}{\theta} = \frac{1}{\theta^n}$:  
$
f(\theta) = \left\{
    \begin{array}\\
        \frac{1}{\theta^n} & \mbox{if } \ X_i \in [0, \theta], i \in [1, n] \\
        0 & otherwise \\
    \end{array}
\right.
$  
Необходимо найти такой $\theta^*$, который удовлетворяет $f(\theta^*) = max_\theta f(\theta)$. Критерий $X_i \in [0, \theta]$ для любого $i$ равносилен для $\theta$ тому, что $\theta \in [max_{1\leq i \leq n}X_i, \infty)$: в выборку из $n$ элементов, **принадлежащих диапазону**, либо попал параметр (тогда max, т.к ряд возрасающий), либо параметр больше, вплоть до бесконечности.  
При этом $max_{1\leq i \leq n}X_i = X_n$ - последний член вариационного ряда (ряд идёт по возрастающей). Итого:  
$
f(\theta) = \left\{
    \begin{array}\\
        \frac{1}{\theta^n} & \mbox{if } \ \theta \in [X_n, \infty) \\
        0 & otherwise \\
    \end{array}
\right.
$  
Функция $\frac{1}{\theta^n}$ убывает на всей области определения, т.о максимум с учётом ограничений достигается при $\hat\theta_{MML} = X_n = max_{1\leq i \leq n}X_i$.

**n2) Покажите, что оценка параметра $θ$ метом моментов есть оценка $\hat{\theta}_2 = 2\bar{X}$**  
Согласно методу моментов необходимо построить функцию от искомого параметра, которая равна $k-му$ моменту. Для 1-го параметра достаточно взять $k=1$, однако если первый момент равен нулю, тогда берём второй, и т.д.  
В нашем случае $E(X_i^1) = \frac{\theta-0}{2} = \frac{\theta}{2}=\bar{X} \Rightarrow \hat{\theta}_{MM} = 2\bar{X}$

## Точечная оценка и асимптотический Д.И.

Судя по гистограмме уместно предположить, что число аварий имеет распределение Пуассона, $Poiss(\lambda)$.  То есть, если мы хотим что-то понять про среднее число аварий, нам надо оценить параметр $\lambda$. 

__Точечная оценка__

Найдите оценку параметра $\lambda$ с помощью метода моментов и метода максимального правдоподобия. Запишите получившийся результат в переменную `lambda_hat`.



$X_1, X_2, ... , X_n \space \tilde{iid} \space Poisson(\lambda)$  
Метод макс. правдоподобия:
* Функция плотности: $P(X=x) = f(x|\lambda) = \dfrac{\lambda ^x e^{-\lambda}}{x!}$, $x = 0,1,2,... \space \lambda > 0$;  
* Функция правдоподобия: $$L(\hat\lambda; x_1, ..., x_n) = \prod^n_{j=1} \frac{\hat\lambda ^x_j e^{-\hat\lambda}}{x_j!}$$  
* Лог функция правдоподобия: $$l(\lambda; x_1, ..., x_n) = ln\begin{pmatrix}
\prod^n_{j=1} \frac{\hat\lambda ^x_j e^{-\hat\lambda}}{x_j!}
\end{pmatrix} = \sum_{j=1}^n[x_j ln\hat\lambda - \hat\lambda -ln(x_j!)] = -n\hat\lambda +ln\hat\lambda \sum_{j=1}^n x_j - \sum_{j=1}^n ln(x_j!)$$  
* Критерий максимизации: $\dfrac{\partial l}{\partial \hat\lambda} = 0$:  
$$\dfrac{\partial l}{\partial \hat\lambda} = -n + \frac{\sum_{j=1}^n x_j}{\hat\lambda} = 0 \Rightarrow \hat\lambda_{MML} = \bar X$$


Метод моментов:  
* Мат. ожидание Пуассоновского распределения $\mathbb{E}[X] = \lambda = \bar X$, откуда $\hat\lambda_{MM} = \bar X$

## Дельта-метод

__Вероятность для распределения Пуассона__ 

Вспомните формулу, по которой считается вероятность того, что Пуассоновская случайная величина примет какое-то значение. Найдите точечную оценку вероятности того, что произойдёт только одна авария. В качестве $\lambda$ используйте оценку метода моментов, полученную в первом задании. 

```python
def poiss_prob(k):
    # your code here
    exp_x = lam_hat**(k)*math.exp(-lam_hat)/math.factorial(k)
    var_x = (lam_hat**k * math.exp(-lam_hat) * (k/lam_hat - 1) / math.factorial(k))**2 * lam_hat/len(df_3)
    return exp_x, var_x

# проверка функции
assert np.abs(poiss_prob(1)[0] - 0.3098761) < 1e-5

```

__Дельта-метод__ 

Та вероятность, которую мы нашли в предыдущем пункте - это лишь точечная оценка. Она является функцией от среднего. С помощью дельта-метода постройте для этой вероятности $95\%$ довертельный интервал.

PMF распределения Пуассона: $\dfrac{\lambda ^ x e^{-\lambda}}{x!}$.  
Её производная  по лямбда: $ \dfrac{\lambda^x e^{-\lambda}\left(\frac{x}{\lambda}-1\right)}{x!} $.

```python
alpha = 0.05
k = 1
ans = poiss_prob(k)

left = stats.norm.interval(1-alpha, loc=ans[0], scale=ans[1]**0.5)[0] #(1-alpha, len(df_3)-1, loc=ans[0], scale=ans[1]**0.5)[0] 
right = stats.norm.interval(1-alpha, loc=ans[0], scale=ans[1]**0.5)[1]
delta = right - left

# проверка, что интервал построен корректно
assert np.abs(left - 0.279) < 1e-3

print(left, right)
```

По аналогии можно построить доверительные интервалы для вероятности того, что случайная величина примет значения от нуля до девяти. Проделайте это в качестве необязательного упражнения и оформите процедуру расчёта в виде функции.

```python
def prob_ci(x, alpha = 0.01, lam_hat=lam_hat):
    # your code here
    left = [stats.norm.interval(1-alpha, loc=poiss_prob(k)[0], scale=poiss_prob(k)[1]**0.5)[0]  for k in x] #stats.t.interval(1-alpha, len(df_3)-1, loc=poiss_prob(k)[0], scale=poiss_prob(k)[1]**0.5)[0]
    right = [stats.norm.interval(1-alpha, loc=poiss_prob(k)[0], scale=poiss_prob(k)[1]**0.5)[1] for k in x]
    return left, right
```

## Д.И. для дисперсии

Предположим, что математическое ожидание результата за контрольную зависит от уровня подготовки, а дисперсия результата — от нервов. Постройте для отношения нервозности плохих и хороших студентов $95\%$ доверительный интервал. Для оценки всех дисперсий используйте несмещённый вариант.  
**PS:** слайд 85, лекция 4.

Вошла ли в доверительный интервал единица? Как думаете, что это означает? 

> *Да. Что плохие студенты *не являются* более нервозными на 5% уровне значимости.*

```python
import scipy
alpha = 0.05
left =  bad.var(ddof=1)/good.var(ddof=1)*scipy.stats.f.ppf(q=alpha/2, dfn=good.size - 1, dfd=bad.size - 1)   # левая граница интервала
right = bad.var(ddof=1)/good.var(ddof=1)*scipy.stats.f.ppf(q=1-alpha/2, dfn=good.size - 1, dfd=bad.size - 1)     # правая граница интервала
delta = right - left     # ширина интервала

# проверка, что границы доверительного интервала найдены верно
assert np.abs(left - 0.6120428247687726) < 1e-5

print(left, right)
```

## Доля написавших третью контрольную

У студентов был выбор: писать или не писать третью контрольную. Если в колонке `k3` стоит $0$, значит студент не стал это делать. Постройте асимптотический $95\%$-й доверительный интервал для разницы долей хороших и плохих студентов, которые решили писать третью контрольную. 

__Примечание:__ На практике, если мы пользуемся асимптотическими доверительными интервалами, мы можем искать границы двумя способами:

* __Первый__ состоит в том, что мы оцениваем стандартное отклонение с помощью $\sqrt{\hat{p} \cdot (1 - \hat{p})} = \sqrt{\bar{x} \cdot (1 - \bar{x})}$. 
* __Второй способ__ состоит в том, чтобы оценить стандартное отклонение как  $\sqrt{\frac{1}{n-1} \cdot \sum_{i=1}^n (x_i - \bar{x})^2}$. 

В плане асимптотики это не влияет на выводы, так как обе этих оценки состоятельно оценивают дисперсию. Можно пользоваться любым.  
**PS:** слайд 55, лекция 3

Входит ли в доверительный интервал ноль? Как думаете, что это означает? 

> *Да. Это говорит, что на уровне 5% опыт списывания не влияет на выбор 3-й контрольной.*

```python
k3_good = df[df['fraud'] == 0]['k3']
k3_bad = df[df['fraud'] > 0]['k3']
p_good = k3_good[k3_good != 0].count() / k3_good.size
p_bad = k3_bad[k3_bad != 0].count() / k3_bad.size
```

```python
alpha = 0.05
left =  p_good - p_bad - stats.norm.ppf(1 - alpha/2)*(p_good*(1-p_good)/k3_good.size + p_bad*(1-p_bad)/k3_bad.size)**0.5   # левая граница интервала
right = p_good - p_bad + stats.norm.ppf(1 - alpha/2)*(p_good*(1-p_good)/k3_good.size + p_bad*(1-p_bad)/k3_bad.size)**0.5    # правая граница интервала
delta = right-left     # ширина интервала

# проверка, что границы доверительного интервала найдены верно
assert np.abs(left + 0.1885) < 1e-3


print(left, right)
```

## Снова разница средних

Посмотрим на результаты студентов в динамике. Постройте точный $95\%$ доверительный интервал для средней разницы оценок за вторую и первую контрольные. Пропуски в данных заполните нулями.  
**PS:** слайд 31, 46, лекция 4.

Входит ли в доверительный интервал ноль? Как думаете, что это означает? 

> *Нет. Доверительный интервал на уровне 5% лежит в отрицательном диапазоне - значит вторая кр была написана в среднем хуже.*

```python
mark_diff = df['k2'].fillna(0) - df['k1'].fillna(0)
mark_diff_mean = mark_diff.mean()
mark_diff_var = mark_diff.var(ddof = 1)
mark_diff
```

Вообще говоря, истинная дисперсия популяции нам неизвестна: есть пропуски по студентам, так или иначе были списывания. Поэтому базовое распределние для ДИ - это t-распределение Стьюдента.

```python
import math
alpha = 0.05
left = mark_diff_mean - stats.t(mark_diff.size-1).ppf(1 - alpha/2)*math.sqrt(mark_diff_var/mark_diff.size)    # левая граница интервала
right = mark_diff_mean + stats.t(mark_diff.size-1).ppf(1 - alpha/2)*math.sqrt(mark_diff_var/mark_diff.size)     # правая граница интервала
delta = right-left     # ширина интервала

# проверка, что границы доверительного интервала найдены верно
assert np.abs(left + 1.4562881732820103) < 1e-5

print(left, right)
```

В предположении об известной истинной дисперсии (пусть пропусков не так много и никто реально не списывал) получим очень близкий к посчитанному выше ДИ (разница в третьем порядке на 1):

```python
left = mark_diff_mean - stats.norm.ppf(1 - alpha/2)*math.sqrt(mark_diff_var/mark_diff.size)    # левая граница интервала
right = mark_diff_mean + stats.norm.ppf(1 - alpha/2)*math.sqrt(mark_diff_var/mark_diff.size)     # правая граница интервала
delta = right-left     # ширина интервала

print(left, right)
```
